**Numerical Differentiation Problem Statement:** We have dealt with regression and interpolation for finding the equation of a function but now we will deal with finding the derivative of a function numerically.

**Finite Difference Approximating Derivatives:** The limit definition of a derivative can be used for approximating the derivative at a point, x. If we take the slope of the line that connects a point, x1, and the next point, x2, this is called the **forward** difference. If we take the slope of the line that connects the points before x and the point at x, then we get the **backward** difference. If we take the slope of the line that connects the points before x and the point after x, then we get the **central** difference. The central difference is more accurate but requires knowing a point before and after point x.


**Approximating of Higher Order Derivatives:** More accurate approximations as well as derivatives of higher degrees can be found by solving a system of linear equation of taylor series of nth degree approximation. These more accurate/higher derivatives require more calculations and points before and after x. A table of these equations sorted by order and accuracy can be found [here](https://en.wikipedia.org/wiki/Finite_difference_coefficient).

```python
import numpy as np
import matplotlib.pyplot as plt

# From the book except using central difference through np.gradient()

# step size
h = 0.1
# define grid
x = np.arange(0, 2*np.pi, h)
# compute function
y = np.cos(x)

# compute vector of forward differences
central_diff = np.gradient(y)/h
central_diff = np.delete(central_diff, len(central_diff)-1, 0)
# compute corresponding grid
x_diff = x[:-1:]
# compute exact solution
exact_solution = -np.sin(x_diff)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff, central_diff, '--', \
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()

# Compute max error between
# numerical derivative and exact solution
max_error = max(abs(exact_solution - central_diff))
print(max_error)
```
